```python
# user-movie-rating-ts
fname = 'ml-100k/u.data'
data_frame = pd.read_csv(fname, sep='\t')
data = data_frame.as_matrix()
```

```python
data = data.copy()[:,:3]
```

```python
data
```

```python
data_matrix = np.zeros([943, 1682])
for r in range(len(data)):
    i, j, rating = list(data[r, :])
    data_matrix[i-1, j-1] = rating 
    
    
```

```python
data_matrix.shape
```

```python
data_matrix -= np.mean(data_matrix)
```

```python
from sklearn.decomposition import PCA

#pca = PCA(n_components=0.8, svd_solver='full')
pca = PCA(svd_solver='full')
```

```python
pca.fit(data_matrix)

pca.n_components_
```

```python
len(pca.components_)
```

```python
from numpy.random import RandomState
from sklearn.datasets import fetch_olivetti_faces

rng = RandomState(0)

# Load faces data
dataset = fetch_olivetti_faces(shuffle=True, random_state=rng)
faces = dataset.data

n_samples, n_features = faces.shape

# global centering
faces_centered = faces - faces.mean(axis=0)

# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)

print("Dataset consists of %d faces" % n_samples)
```

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=0.8, svd_solver='full')
```

```python
import os
tot_n_images = 0
for (dirpath, _, _) in os.walk('aligned'):
    tot_n_images += len(os.listdir(dirpath))
print(f"Total number of aligned images is {tot_n_images}")
```

```python
import cv2

img_data = []

for (dirname, _, filenames) in os.walk('aligned'):
    for im in filenames:
        im_path = os.path.join(dirname, im)
        grey_img = cv2.imread(im_path, 0)
        img_data.append(grey_img.flatten())
        
```
